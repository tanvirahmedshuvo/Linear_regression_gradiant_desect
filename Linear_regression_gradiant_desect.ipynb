```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x, y):
    m_curr = b_curr = 0  # Initial values for slope and intercept
    learning_rate = 0.01  # Learning rate
    n = len(x)  # Number of data points
    iterations = 10000  # Number of iterations

    plt.scatter(x, y, color='red', marker='+', linewidth=5)  # Scatter plot of data

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr  # Hypothesis
        cost = (1/n) * sum((y - y_predicted) ** 2)  # Cost function (MSE)
        
        md = -(2/n) * sum(x * (y - y_predicted))  # Partial derivative w.r.t m
        bd = -(2/n) * sum(y - y_predicted)  # Partial derivative w.r.t b
        
        m_curr -= learning_rate * md  # Update m
        b_curr -= learning_rate * bd  # Update b
        print(f"Iteration {i}: m={m_curr}, b={b_curr}, cost={cost}")

    # Plot the final regression line
    plt.plot(x, m_curr * x + b_curr, color='green')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent Regression Line')
    plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
